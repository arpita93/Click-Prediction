```python
import pandas as pd

from sklearn.feature_extraction import FeatureHasher

import gc

import tensorflow as tf

from sklearn import preprocessing

from sklearn.preprocessing import Imputer

import numpy as np
from sklearn.preprocessing import StandardScaler


def normalize_data(df):
    
    
    imp = Imputer(missing_values='NaN', strategy='mean', axis=0)
    
    scaled_values = imp.fit_transform(df)
    
    scaler = StandardScaler() 
    scaled_values = scaler.fit_transform(scaled_values)
    
    
    
    return scaled_values
    

def encoder_data(train_nrows,test_nrows,cols):
    
    train_c = np.array([]).reshape(train_nrows,0)
    test_c = np.array([]).reshape(test_nrows,0)
    

    for c in cols:
        
                
        train_t_df = pd.read_csv("train",usecols=[c],dtype=object).astype('str')
        test_t_df = pd.read_csv("test",usecols=[c],dtype=object).astype('str')
        
        union = pd.concat([train_t_df,test_t_df], ignore_index=True).values
        
        h = FeatureHasher(n_features=10,input_type="string")
        
        union_fit = h.transform(union)

        train_fit = union_fit[test_nrows:].toarray()
        test_fit = union_fit[:test_nrows].toarray()
        
        
        train_c = np.concatenate ((train_c,train_fit), axis = 1)
        test_c = np.concatenate ((test_c,test_fit) , axis = 1)
        
        
                
    return train_c, test_c


print "Start Reading"

#'n12'
ncols = ['n1','n2','n3','n4','n5','n6','n7','n8','n9','n10','n11','n12','n13']

ccols = ['c2','c3','c4','c5','c6','c7','c8','c9','c10','c11','c12'\
                                          ,'c13','c14','c15','c16','c17','c18','c19','c20','c21','c22','c23','c24'\
                                         ,'c25','c26','c27']


train_n_df = pd.read_csv("train",usecols=ncols)

train_n_df['n12'] = train_n_df['n12'].str.replace('Unnamed: ', '')


test_n_df = pd.read_csv("test",usecols=ncols)

train_click_df = pd.read_csv("train",usecols=['click'])
test_click_df = pd.read_csv("test",usecols=['click'])


print "Done Reading"



train_y = pd.DataFrame(data= {'0': np.where(train_click_df['click']== 0 , 1, 0) , '1': np.where(train_click_df['click'] == 1 , 1, 0) })

test_y = pd.DataFrame(data= {'0': np.where(test_click_df['click']== 0 , 1, 0) , '1': np.where(test_click_df['click'] == 1 , 1, 0) })


#test_id = test_id_df["id"]
print "Start Preprocessing"


train_n =  normalize_data(train_n_df)
test_n =  normalize_data(test_n_df)


train_c , test_c = encoder_data(len(train_click_df),len(test_click_df),ccols)


train_s = np.concatenate ((train_n,train_c,train_y), axis = 1)
test_s = np.concatenate ((test_n,test_c,test_y), axis = 1)



np.savetxt('train_p',train_s, fmt='%f', delimiter=',')
np.savetxt('test_p', test_s, fmt='%f', delimiter=',')


print "End Preprocessing"








```
